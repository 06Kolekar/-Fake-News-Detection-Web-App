```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23480,23470,-1):
    data_fake.drop([i], axis = 0 , inplace = True)

data_true_manual_testing = data_true.tail(10)
for i in range(21416,21406,-1):
    data_true.drop([i], axis = 0 , inplace = True)
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing['classs'] = 0
data_true_manual_testing['classs'] = 1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake , data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title','subject','date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'],axis = 1 , inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '',text)
    text = re.sub('\n', '',text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test , y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train , y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
print(classification_report(y_test, pred_rf))
```

```python
import pandas as pd

def output_label(n):
    return "Fake News" if n == 0 else "Real News"

def manual_testing():
    # Take input from the user
    raw_news = input("Enter news content: ")
    
    # Preprocess the input using wordopt
    cleaned_news = wordopt(raw_news)
    
    # Debug: see what the cleaned text looks like
    print("\nCleaned Input:", cleaned_news)
    
    # Convert into DataFrame and apply vectorizer
    test_df = pd.DataFrame({"text": [raw_news], "test": [cleaned_news]})
    vectorized_input = vectorization.transform(test_df["test"])
    
    # Predict using multiple models
    pred_LR = LR.predict(vectorized_input)
    pred_DT = DT.predict(vectorized_input)
    pred_GB = GB.predict(vectorized_input)
    pred_RF = RF.predict(vectorized_input)
    
    # Output predictions
    print("\n--- Model Predictions ---")
    print("Logistic Regression: ", output_label(pred_LR[0]))
    print("Decision Tree       : ", output_label(pred_DT[0]))
    print("Gradient Boosting   : ", output_label(pred_GB[0]))
    print("Random Forest       : ", output_label(pred_RF[0]))

# Run the prediction
manual_testing()
```
